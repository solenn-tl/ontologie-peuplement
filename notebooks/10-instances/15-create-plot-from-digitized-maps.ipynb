```python
#Cr√©er un objet Turtle avec une URI
g = Graph()

g.bind('landmark', landmarkuri)
g.bind('owner', owneruri)
g.bind('source', srcuri)
g.bind('change', changeuri)
g.bind('event',eventuri)
g.bind('cad_ltype', cad_ltype)
g.bind('cad_atype', cad_atype)
g.bind('pnature',pnature)
g.bind('ctype',ctype)
g.bind('atype',atype)

g.bind('cad', cad)
g.bind('add', add)
g.bind('rico', rico)
g.bind('fpo', fpo)
g.bind('time',time)
```

```python
def get_base_Skos_Concept(str):
    if str == 'B':
        return 'ProprieteBatie'
    elif str == 'NB':
        return 'ProprieteNonBatie'
    elif str == 'B-NB' or str == 'NB-B':
        return 'ProprieteBatie;ProprieteNonBatie'
    
def geojson_to_wkt(geojson_geometry):
    shapely_geometry = shape(geojson_geometry)
    return shapely_geometry.wkt
```

```python
geojsons = [data_1811, data_1845]
for i in range(len(geojsons)):
    if i == 0:
        CAD = '1811'
        BEGIN = '1811-01-01'
        END = '1848-03-31'
        LANDMARK_SECTION = "87d7c2f6-306b-45a1-a833-5e17821c3102"
    elif i == 1:
        CAD = '1845'
        BEGIN = '1845-01-01'
        END = '1860-12-31'
        LANDMARK_SECTION = 'da6a5c2c-e86d-43bb-8950-7169bd0df60a'

    for elem in geojsons[i]['features']:

        #Create Landmark of type Plot
        resource_uri = URIRef(landmarkuri + NUM_DEPT + '_' + COMMUNE + '_' +  CAD + '_' + elem['properties']['section'] + str(elem['properties']['numero']) + '_' + str(uuid.uuid4()))
        g.add((resource_uri, RDF.type, add.Landmark))
        g.add((resource_uri, add.isLandmarkType, cad_ltype.Plot))
        g.add((resource_uri, DCTERMS.identifier, Literal(elem['properties']['section'] + '-' + elem['properties']['numero'], datatype=XSD.string)))
        g.add((resource_uri, SKOS.prefLabel, Literal(elem['properties']['section'] + '-' + str(elem['properties']['numero']) + ' ('+ COMMUNE +')', datatype=XSD.string)))
        g.add((resource_uri, RDFS.label, Literal(elem['properties']['section'] + '-' + str(elem['properties']['numero']) + ' ('+ COMMUNE +')', datatype=XSD.string)))

        #Create Plot Landmark Relation : Plot <=> Section
        landmarkRelation = URIRef(BNode().n3())
        g.add((landmarkRelation, RDF.type, add.LandmarkRelation))
        g.add((landmarkRelation, add.locatum, resource_uri))
        g.add((landmarkRelation, add.relatum, URIRef(landmarkuri + LANDMARK_SECTION)))
        g.add((landmarkRelation, add.isLandmarkRelationType, lrtype.Within))

        #Landmark Appearance
        ## Change
        changeapp = URIRef(changeuri + str(uuid.uuid4()))
        g.add((changeapp, RDF.type, add.Change))
        g.add((changeapp,add.appliedTo,resource_uri))
        g.add((changeapp,add.isChangeType,ctype.LandmarkAppearance))
        
        ## Event
        eventapp = URIRef(eventuri + str(uuid.uuid4()))
        g.add((eventapp, RDF.type, add.Event))
        g.add((changeapp,add.dependsOn,eventapp))
        time_ = BNode()
        g.add((eventapp, add.hasTime, URIRef(time_.n3())))
        g.add((URIRef(time_.n3()), RDF.type, add.TimeInstant))
        g.add((URIRef(time_.n3()), add.timeCalendar,time.Gregorian))
        g.add((URIRef(time_.n3()), add.timePrecision,time.Year))
        g.add((URIRef(time_.n3()), add.timeValue, Literal(datetime.strptime(BEGIN, '%Y-%m-%d'), datatype=XSD.dateTimeStamp)))

        #Landmark Disappearance (NOT USED)
        ## Change
        #changedis = URIRef(changeuri + str(uuid.uuid4()))
        #g.add((changedis, RDF.type, add.Change))
        #g.add((changedis,add.appliedTo,resource_uri))
        #g.add((changedis,add.isChangeType,ctype.LandmarkDisappearance))
        
        ## Event
        #eventdis= URIRef(eventuri + str(uuid.uuid4()))
        #g.add((eventdis, RDF.type, add.Event))
        #g.add((changedis,add.dependsOn,eventdis))
        #time_ = BNode()
        #g.add((eventdis, add.hasLatestTimeInstant, URIRef(time_.n3())))
        #g.add((URIRef(time_.n3()), RDF.type, add.TimeInstant))
        #g.add((URIRef(time_.n3()), add.timeCalendar,time.Gregorian))
        #g.add((URIRef(time_.n3()), add.timePrecision,time.Year))
        #g.add((URIRef(time_.n3()), add.timeValue, Literal(datetime.strptime(END, '%Y-%m-%d'), datatype=XSD.dateTimeStamp)))

        # Plot Nature
        ## Version
        natureattr = BNode()
        g.add((URIRef(natureattr.n3()),RDF.type,add.Attribute))
        g.add((resource_uri, add.hasAttribute, URIRef(natureattr.n3())))
        g.add((URIRef(natureattr.n3()), add.isAttributeType, cad_atype.PlotNature))
        natureattrversion = BNode()
        g.add((URIRef(natureattr.n3()), add.hasAttributeVersion,URIRef(natureattrversion.n3())))
        g.add((URIRef(natureattrversion.n3()),RDF.type,add.AttributeVersion))
        nature = get_base_Skos_Concept(elem['properties']['nature'])
        if ';' in nature:
            for n in nature.split(';'):
                g.add((URIRef(natureattrversion.n3()),cad.hasPlotNature,URIRef(pnature+n)))
        else:
            g.add((URIRef(natureattrversion.n3()),cad.hasPlotNature,URIRef(pnature+nature)))
        
        ## Plot Nature Appearance
        ### Change
        changenatapp = URIRef(changeuri + str(uuid.uuid4()))
        g.add((changenatapp, RDF.type, add.Change))
        g.add((changenatapp, add.isChangeType, ctype.AttributeVersionAppearance))
        g.add((changenatapp,add.appliedTo,URIRef(natureattr.n3())))
        g.add((changenatapp,add.makesEffective,URIRef(natureattrversion.n3())))
        ### Event
        eventnatapp = URIRef(eventuri + str(uuid.uuid4()))
        g.add((eventnatapp, RDF.type, add.Event))
        g.add((changenatapp,add.dependsOn,eventnatapp))
        time_ = BNode()
        g.add((eventnatapp, add.hasTime, URIRef(time_.n3())))
        g.add((URIRef(time_.n3()), RDF.type, add.TimeInstant))
        g.add((URIRef(time_.n3()), add.timeCalendar,time.Gregorian))
        g.add((URIRef(time_.n3()), add.timePrecision,time.Year))
        g.add((URIRef(time_.n3()), add.timeValue, Literal(datetime.strptime(BEGIN, '%Y-%m-%d'), datatype=XSD.dateTimeStamp)))

        ## Plot Nature Disappearance
        ### Change
        #changenatdis = URIRef(changeuri + str(uuid.uuid4()))
        #g.add((changenatdis, RDF.type, add.Change))
        #g.add((changenatdis,add.appliedTo,URIRef(natureattr.n3())))
        #g.add((changenatdis,add.outdates,URIRef(natureattrversion.n3())))
        ### Event
        #eventnatdis = URIRef(eventuri + str(uuid.uuid4()))
        #g.add((eventnatdis, RDF.type, add.Event))
        #g.add((changenatdis,add.dependsOn,eventnatdis))
        #time_ = BNode()
        #g.add((eventnatdis, add.hasLatestTimeInstant, URIRef(time_.n3())))
        #g.add((URIRef(time_.n3()), RDF.type, add.TimeInstant))
        #g.add((URIRef(time_.n3()), add.timeCalendar,time.Gregorian))
        #g.add((URIRef(time_.n3()), add.timePrecision,time.Year))
        #g.add((URIRef(time_.n3()), add.timeValue, Literal(datetime.strptime(END, '%Y-%m-%d'), datatype=XSD.dateTimeStamp)))

        #Plot Geometry
        geom = BNode()
        g.add((URIRef(geom.n3()), RDF.type, add.Attribute))
        g.add((URIRef(geom.n3()), add.isAttributeType, atype.Geometry))
        g.add((resource_uri, add.hasAttribute, URIRef(geom.n3())))
        geomversion = BNode()
        g.add((URIRef(geom.n3()), add.hasAttributeVersion, URIRef(geomversion.n3())))
        g.add((URIRef(geomversion.n3()), RDF.type, add.AttributeVersion))
        g.add((URIRef(geomversion.n3()), add.versionValue, Literal("<http://www.opengis.net/def/crs/OGC/1.3/CRS84>" + geojson_to_wkt(elem['geometry']), datatype=geo.wktLiteral)))
        g.add((URIRef(geomversion.n3()), PROV.wasGeneratedBy, URIRef(activity + "0006")))

        ## Plot Geometry Appearance
        ### Change
        changegeomapp = URIRef(changeuri + str(uuid.uuid4()))
        g.add((changegeomapp, RDF.type, add.Change))
        g.add((changegeomapp, add.isChangeType, ctype.AttributeVersionAppearance))
        g.add((changegeomapp, add.appliedTo, URIRef(geom.n3())))
        g.add((changegeomapp, add.makesEffective, URIRef(geomversion.n3())))
        ### Event
        eventgeomapp = URIRef(eventuri + str(uuid.uuid4()))
        g.add((eventgeomapp, RDF.type, add.Event))
        g.add((changegeomapp, add.dependsOn, eventgeomapp))
        time_ = BNode()
        g.add((eventgeomapp, add.hasTime, URIRef(time_.n3())))
        g.add((URIRef(time_.n3()), RDF.type, add.TimeInstant))
        g.add((URIRef(time_.n3()), add.timeCalendar, time.Gregorian))
        g.add((URIRef(time_.n3()), add.timePrecision, time.Year))
        g.add((URIRef(time_.n3()), add.timeValue, Literal(datetime.strptime(BEGIN, '%Y-%m-%d'), datatype=XSD.dateTimeStamp)))

        ## Plot Geometry Disappearance
        ### Change
        #changegeomdis = URIRef(changeuri + str(uuid.uuid4()))
        #g.add((changegeomdis, RDF.type, add.Change))
        #g.add((changegeomdis, add.appliedTo, URIRef(geom.n3())))
        #g.add((changegeomdis, add.outdates, URIRef(geomversion.n3())))
        ### Event
        #eventgeomdis = URIRef(eventuri + str(uuid.uuid4()))
        #g.add((eventgeomdis, RDF.type, add.Event))
        #g.add((changegeomdis, add.dependsOn, eventgeomdis))
        #time_ = BNode()
        #g.add((eventgeomdis, add.hasLatestTimeInstant, URIRef(time_.n3())))
        #g.add((URIRef(time_.n3()), RDF.type, add.TimeInstant))
        #g.add((URIRef(time_.n3()), add.timeCalendar, time.Gregorian))
        #g.add((URIRef(time_.n3()), add.timePrecision, time.Year))
        #g.add((URIRef(time_.n3()), add.timeValue, Literal(datetime.strptime(END, '%Y-%m-%d'), datatype=XSD.dateTimeStamp)))

print(g.serialize(format='turtle'))
g.serialize(destination=f"{ROOT}data/rdf/{COMMUNE}_landmarks_initial_plots.ttl", format='turtle')
```
