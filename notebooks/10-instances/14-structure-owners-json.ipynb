```python
prompt_demand = """Return the response as a JSON file formatted like this (number of elements in owners and changes keys depend on the input text) : 
Key names should with no backslashes, values should use plain ascii with no special characters and all elements should be in the same list.'
"""
```

```python
def entry_to_prompt(entry):
    prompt = []
    prompt.append("Input:")
    prompt.append(json.dumps(dict(registre=entry["registre"],type_folio=entry["Type_CF"],folio=entry["Num_Folio"],groupe_cf=entry["Groupe CF"],transcription=entry["Propriétaires_transcript"]),indent = 2, ensure_ascii=False, default=str))
    prompt.append("Output:")

    return "\n".join(prompt).strip()

print(entry_to_prompt(owners.iloc[11]))
```

```python
def few_shot_prompt(df,indexs):
    few_shot_prompt = []

    for index, entry in df.iterrows():
        if index in indexs:
            few_shot_prompt.append("\nInput:")
            few_shot_prompt.append(json.dumps(dict(registre=entry["registre"],type_folio=entry["Type_CF"],folio=entry["Num_Folio"],groupe_cf=str(int(entry["Groupe CF"])),transcription=entry["Propriétaires_transcript"]),indent = 2, ensure_ascii=False, default=str))
            few_shot_prompt.append("Output:")
            few_shot_prompt.append(entry["JSON"])

    return "\n".join(few_shot_prompt).strip()

#print(few_shot_prompt(train,[1,3,4,15,18,24,25]))
```
