## Create few-shots

````python
import ollama
import json
````

````python
prompt_intro = """
You are a digital humanities research assistant. 
Your goal is to structure cells of a table that described one ore many land owners of a list of parcels. 
Owner cells have been manually copied from archival records from the 19th century. Texts are in French.

There are special tokens in the text : 
~~TEXT~~ : is markdown syntaxe to say that some part of the text is stripped in the original record
→ : means that there is a line break in the original record
↑TEXT↓ : means that the text superscript in the original record

Here are some examples of inputs and their associated JSON outputs :
"""
````

````python
prompt_demand = """For the given JSON intput, please produce a JSON output that is structured as the following examples. 
ONLY RETURN A VALID JSON (output) AS RESPONSE and nothing else. do.
"""
````

````python
def entry_to_prompt(entry):
    prompt = []
    prompt.append("Input:")
    prompt.append(json.dumps(dict(registre=entry["registre"],type_folio=entry["Type_CF"],folio=entry["Num_Folio"],groupe_cf=entry["Groupe CF"],transcription=entry["Propriétaires_transcript"]),indent = 2, ensure_ascii=False, default=str))
    prompt.append("Output:")

    return "\n".join(prompt).strip()

print(entry_to_prompt(owners.iloc[11]))
````

````python
def few_shot_prompt(df,indexs):
    few_shot_prompt = []

    for index, entry in df.iterrows():
        if index in indexs:
            few_shot_prompt.append("\nInput:")
            few_shot_prompt.append(json.dumps(dict(registre=entry["registre"],type_folio=entry["Type_CF"],folio=entry["Num_Folio"],groupe_cf=str(int(entry["Groupe CF"])),transcription=entry["Propriétaires_transcript"]),indent = 2, ensure_ascii=False, default=str))
            few_shot_prompt.append("Output:")
            few_shot_prompt.append(entry["JSON"])

    return "\n".join(few_shot_prompt).strip()

#print(few_shot_prompt(train,[1,3,4,15,18,24,25]))
````

## Run

````python
from ollama import Client, generate, embeddings
ollama = Client(host='127.0.0.1:11434')
````

````python
base_prompt = prompt_intro + "\n" + few_shot_prompt(train,[1,2,5,13,16,21]) + '\n' + prompt_demand + '\n'
base_prompt
````

````python
import json
import re
def clean_answer(text):
    pattern = r'{(.*)}'
    text = text.replace('\n','')
    match = re.search(pattern, text)
    res = match.group(0)
    res = json.loads(res)
    return res
````

````python
structured_owners = []
for index,row in owners.iterrows():
    full_prompt = base_prompt + entry_to_prompt(owners.iloc[index])
    answer = ollama.generate(model='llama3',prompt=full_prompt)
    structured_owners.append(answer)
````

````python
list_cf_ok = []
list_cf_nok = []
count = 0
for elem in structured_owners:
    try:
        list_cf_ok.append(clean_answer(elem['response']))
    except:
        list_cf_nok.append(elem['response'])
        print(count)
    count+=1
````
