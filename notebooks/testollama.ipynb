## Create few-shots

```python
import ollama
import json
import ast
```

```python
def few_shot_prompt(df,indexs):
    few_shot_prompt = []

    for index, entry in df.iterrows():
        if index in indexs:
            few_shot_prompt.append("\nInput:")
            few_shot_prompt.append(json.dumps(dict(registre=entry["registre"],type_folio=entry["Type_CF"],folio=entry["Num_Folio"],groupe_cf=["Groupe CF"],transcription=["Propriétaires_transcript"]),indent = 2, ensure_ascii=False))
            few_shot_prompt.append("Analysis:")
            few_shot_prompt.append(entry["JSON"])

    return "\n".join(few_shot_prompt).strip()

print(few_shot_prompt(train,[1,3,4,15,18,24,25]))
```
