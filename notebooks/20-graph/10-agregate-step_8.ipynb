```python
PATH = 'C:/Users/stual/PycharmProjects/peuplement_ontology/ontologie-peuplement/data/rdf-graph-construct/'
CSV = PATH + 'step_8_list_of_nodes.csv'

import pandas as pd

df = pd.read_csv(CSV)
res = []
for i, row in df.iterrows():
    elems = row[0].split(' ')
    elems = sorted(elems)
    res.append(elems)

display(sorted(res))
```

```python
def remove_duplicate_lists(list_of_lists):
    # Convert inner lists to tuples
    list_of_tuples = [tuple(lst) for lst in list_of_lists]
    
    # Use a set to remove duplicates
    unique_tuples = list(set(list_of_tuples))
    
    # Convert the tuples back to lists
    unique_lists = [list(tpl) for tpl in unique_tuples]
    
    return unique_lists
```

```python
unique_lists = remove_duplicate_lists(sorted(res))
unique_lists
```
